```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Generate values for the x-axis
x = np.linspace(-10, 10, 100)

# Calculate corresponding y values using the sigmoid function
y = sigmoid(x)
```

```python
print(x)
```

```python
# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0.5, color='red', linestyle='--', label='y = 0.5')
plt.axvline(0, color='green', linestyle='--', label='x = 0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Step function with Vectorized implementation
def step_function(x):
    ''' 
    if x >= 0:
        return 1
    else:
        return 0
    '''
    return np.where(x >= 0, 1, 0)
```

```python
# Generate values for the x-axis
x = np.linspace(-5, 5, 100)

# Calculate corresponding y values using the step function
y = step_function(x)
```

```python
# Plot the step function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Step Function')
plt.title('Step Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0.5, color='red', linestyle='--', label='f(x) = 0.5')
plt.axvline(0, color='green', linestyle='--', label='x = 0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
# Generating x values
x = np.linspace(-10, 10, 100)  # Creating 200 points between -7 and 7

# Calculate sigmoid function values
sigmoid_y = sigmoid(x)

# Calculate sigmoid derivative values
sigmoid_derivative_y = sigmoid_derivative(x)
```

```python
# Plotting sigmoid function's derivative
plt.subplot(2, 1, 2)
plt.title('Sigmoid Function Derivative')
plt.plot(x, sigmoid_derivative_y, label='Sigmoid Derivative', color='red')
plt.xlabel('x')
plt.ylabel('sigmoid\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
